```python
# Code example / demonstration
1Ô∏è‚É£ Creating lists (Code cell)

numbers = [1, 2, 3, 4, 5]
languages = ["Python", "Java", "C++"]
mixed = [1, "Python", True]

print(numbers)
print(languages)
print(mixed)
```

```python
2Ô∏è‚É£ Indexing lists (Code cell)

languages = ["Python", "Java", "C++"]

print(languages[0])   # 'Python'
print(languages[1])   # 'Java'
print(languages[-1])  # 'C++'
```

```python
3Ô∏è‚É£ Slicing lists (Code cell)

print(languages[0:2])   # ['Python', 'Java']
print(languages[1:])    # ['Java', 'C++']
```
